```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "metre_kare": [700, 2538, 1836, 1836, 4234, 3471, 2475, 4597, 717, 2749],
    "reklam_tarihi": ["10.05.2023", "04.05.2023", "12.05.2023", "12.05.2023", "14.05.2023", "15.05.2023", "10.06.2023", None, "06.06.2023", "15.06.2023"],
    "reklam_fiyati": [400000.0, 1200000.0, 853000.0, 853000.0, 3000500.0, 4250000.0, 2550000.0, 4598000.0, 12000000.0, None],
    "satis_tarihi": ["1.6.2023", "8.5.2023", "30.5.2023", "30.5.2023", "1.6.2023", "29.6.2023", "4.6.2023", "15.6.2023", None, None],
    "satis_fiyati": [400000.0, 1215000.0, 850000.0, 850000.0, 3001000.0, 4000000.0, 2550000.0, 12000000.0, 5000000.0, None],
    "kategori": ["Daire", "Müstakil", "Müstakil", "Müstakil", None, "Müstakil", "Müstakil", "Daire", None, None],
    "Alici": ["Murat", "Ayşe", "Selen", "Serdar", "Ayhan", "Ferdi", "Tülay", "Selin", "Ahmet", None],
    "Bolge": [None]*10
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["reklam_tarihi"] = df["reklam_tarihi"].fillna("10.05.2023")
df["satis_tarihi"] = df["satis_tarihi"].fillna("01.06.2023")
```

```python
df["reklam_fiyati"] = df["reklam_fiyati"].fillna(df["reklam_fiyati"].mean())
df["satis_fiyati"] = df["satis_fiyati"].fillna(df["satis_fiyati"].mean())
# Sayısal değerleri ortalama ile doldurdum.
```

```python
df["kategori"] = df["kategori"].fillna("Bilinmiyor")
df["Alici"] = df["Alici"].fillna("Bilinmiyor")
# Kategorik değerleri "Bilinmiyor" ile doldurdum
```

```python
df["reklam_tarihi"] = pd.to_datetime(df["reklam_tarihi"], dayfirst=True)
df["satis_tarihi"] = pd.to_datetime(df["satis_tarihi"], dayfirst=True)
# Tarih sütunlarını datetime formatına çevirme işlemini yaptım
```

```python
# Outlier tespiti ve silinmesi kısmı (IQR yöntemi ile yaptım)
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    alt_sinir = Q1 - 1.5 * IQR
    ust_sinir = Q3 + 1.5 * IQR
    return df[(df[column] >= alt_sinir) & (df[column] <= ust_sinir)]

# Aykırı değerlerden arındırma işlemi burada
for col in ["metre_kare", "reklam_fiyati", "satis_fiyati"]:
    df = remove_outliers(df, col)
```

```python
if "Bolge" in df.columns:
    df = df.drop("Bolge", axis=1) # İki kez çalıştırdığım için hata verdi o sebeple koşul kontrolü yaptım.
```

```python
for col in ["kategori", "Alici"]:
    if col in df.columns:
        df = pd.get_dummies(df, columns=[col], drop_first=True)
# Kategorik sütunları one-hot encoding ile dönüştürmeye yarar pd.get_dummies fonksiyonu
```
