**Seaborn Taxi Veri seti ile**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("taxis.csv")
df
```

1- Veri Setinin Satır ve Sütun sayısı:

```python
df.shape
```

2- Her bir sütundaki null değerleri getir

```python
df.isnull().sum()
```

3- Her Bir Sütunun veri tipini getir

```python
df.dtypes
```

4- Bahşiş (tip) dağılımını getirin

```python
sns.histplot(df["tip"], bins=30, kde=True)
plt.xlabel("Bahşiş (Tip) Miktarı")
plt.ylabel("Frekans")
plt.title("Bahşiş Dağılımı")
plt.show()
```

5- Biniş yeri (pickup_borough) "Manhattan" olan kayıtları getir

```python
manhattan_trips = df[df["pickup_borough"] == "Manhattan"]
manhattan_trips
```

6- İlçeye göre (pickup_borough) yeşil ve sarı taksilerin ortalama mesafelerini getirin

```python
avg_distance_by_borough = df.groupby(["pickup_borough", "color"])["distance"].mean().unstack()
avg_distance_by_borough
```

7- 0.5 milden küçük (distance) kayıtları getirin

```python
short_trips = df[df["distance"] < 0.5]
short_trips
```

8- Mesafe ile Bahşiş arasındaki dağılımı getirin

```python
sns.scatterplot(x=df["distance"], y=df["tip"])
plt.xlabel("Mesafe (mil)")
plt.ylabel("Bahşiş (tip)")
plt.title("Mesafe ve Bahşiş Dağılımı")
plt.show()
```

9- Her bir alanın grafiğini çizdirin

```python
df.hist(figsize=(12, 8), bins=30)
plt.suptitle("Tüm Sütunların Histogramları", fontsize=16)
plt.show()
```

10- Her bir ilçeye göre (pickup_borough) toplam ortalama değerleri getirin (grafik)

```python
avg_by_borough = df.groupby("pickup_borough").mean(numeric_only=True)
avg_by_borough.plot(kind="bar", figsize=(10,5))
plt.xlabel("İlçe")
plt.ylabel("Ortalama Değerler")
plt.title("İlçeye Göre Ortalama Değerler")
plt.show()
```

11- Her bir ilçenin (pickup_borough) sayı grafiğini çizdirin

```python
sns.countplot(x=df["pickup_borough"])
plt.xlabel("İlçe")
plt.ylabel("Seyahat Sayısı")
plt.title("İlçelere Göre Seyahat Sayıları")
plt.xticks(rotation=45)
plt.show()
```

12- Her bir ilçeye göre sarı ve yeşil taksilerle yapılan seyehatleri getirin

```python
taxi_counts = df.groupby(["pickup_borough", "color"]).size().unstack()
taxi_counts
```

13- "total" alanının kutu grafiğini getirin

```python
sns.boxplot(x=df["total"])
plt.xlabel("Toplam Ücret (total)")
plt.title("Total Değerlerinin Kutu Grafiği")
plt.show()
```

14- İlçe bazında "total" kutu grafiklerini getirin

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df["pickup_borough"], y=df["total"])
plt.xlabel("İlçe")
plt.ylabel("Toplam Ücret (total)")
plt.title("İlçeye Göre Toplam Ücret Dağılımı")
plt.xticks(rotation=45)
plt.show()
```

15- Keman grafiğini oluşturun

```python
plt.figure(figsize=(10,6))
sns.violinplot(x=df["pickup_borough"], y=df["total"])
plt.xlabel("İlçe")
plt.ylabel("Toplam Ücret (total)")
plt.title("İlçeye Göre Total Değerlerinin Keman Grafiği")
plt.xticks(rotation=45)
plt.show()
```

16- striplot çizdir

```python
plt.figure(figsize=(10,6))
sns.stripplot(x=df["pickup_borough"], y=df["total"], jitter=True)
plt.xlabel("İlçe")
plt.ylabel("Toplam Ücret (total)")
plt.title("Stripplot: İlçeye Göre Total Dağılımı")
plt.xticks(rotation=45)
plt.show()
```

17- Swarm plot

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x=df["pickup_borough"], y=df["total"])
plt.xlabel("İlçe")
plt.ylabel("Toplam Ücret (total)")
plt.title("Swarm Plot: İlçeye Göre Total Dağılımı")
plt.xticks(rotation=45)
plt.show()
```

18- İlçeye göre ödeme yöntemlerini ızgara grafikler olarak gösterin (payment, pickup_borough, map: fare)

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.groupby(['pickup_borough', 'payment'])['fare'].mean().unstack(), annot=True, fmt=".2f", cmap="viridis")
plt.title('İlçelere Göre Ödeme Yöntemlerine Göre Ortalama Ücret')
plt.xlabel('Ödeme Yöntemi')
plt.ylabel('İlçe')
plt.show()
```
