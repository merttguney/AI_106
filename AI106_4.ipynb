```python
data = {
    'Kategori': ['Elektronik', 'Giyim', 'Giyim', 'Elektronik', 'Elektronik', 'Giyim'],
    'Marka': ['Samsung', 'Nike', 'Adidas', 'Apple', 'Sony', 'Levis'],
    'Fiyat': [2000, 300, 250, 5000, 1500, 150]
}

df = pd.DataFrame(data)
df
```

```python
grouped = df.groupby('Kategori').agg({'Fiyat': ['mean', 'min', 'max']})
print(grouped)
```

```python
grouped = df.groupby(['Kategori',"Marka"]).agg({'Fiyat': ['mean', 'min', 'max']})
print(grouped)
```

```python
durum = False
kelimeler = []

while durum == False:
      cevap = input('Bir isim giriniz: ')

      if cevap.lower() == 'tamam':
            durum = True
      else:
            kelimeler.append(cevap)
```

```python

kelimeler = []

while True:
  kelime = input("Bir seyler yaziniz: ")

  if kelime == "Tamam":
    print("Tamamlandi.")
    break # Burada 'Tamam' kelimesini eklememesi için döngüyü sonlandırdım.

  kelimeler.append(kelime)

print(kelimeler)
```

```python
from random import randint

tahmin = False
sayi = randint(0, 100)
tahmin_sayisi = 0

while not tahmin:
      cevap = input('Tuttuğum sayıyı tahmin et ')
      tahmin_sayisi += 1

      if int(cevap) == sayi:
            print('Tebrikler  Bildin !!!')
            print(' {}. Tahminde buldun !'.format(tahmin_sayisi))
            break
      elif int(cevap) > sayi:
           print('Sayı {} daha küçük.'.format(cevap))
      elif int(cevap) < sayi:
           print('Sayı {} daha büyük.'.format(cevap))
```
